# Setup colab-utils repo
Since this repo has the convenience functions needed to intercat with git, it can't be used on itself so this notebook manually walks through the installation/updating process.



```python
import os
import json
```

```python
# Connect to google drive
from google.colab import drive
%cd /content
drive.mount('/content/gdrive')
```

```python
# Load git credentials into environment variables
PARENT_FOLDER = '/content/gdrive/MyDrive/repos'
path=f"{PARENT_FOLDER}/env.json"
with open(path) as f:
    env = json.load(f)
for key,value in env.items():
    os.environ[key]=value
```

```python
# Set git config
os.system(f'git config --global user.name  "{os.environ["GH_NAME"]}"')
os.system(f'git config --global user.email "{os.environ["GH_EMAIL"]}"')
```

## Clone repo (the first time)
Do this if you haven't yet cloned the repo

```python
# navigate to parent directory on Google Drive (where repos live)
os.chdir(PARENT_FOLDER)

# clone the repo
git_repo = 'trevorki/colab-utils'
git_str = f"https://{os.environ['GH_TOKEN']}@github.com/{git_repo}.git"
!git clone {git_str}
```

The repo is now set up and ready to use.
